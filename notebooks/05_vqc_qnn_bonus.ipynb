# 05 — Bonus: VQC and QNN

Beyond the kernel-method QSVM, two parameterized-circuit models:

* **VQC** (Variational Quantum Classifier) — `ZZFeatureMap` input layer,
  `RealAmplitudes` ansatz, COBYLA optimizer.
* **QNN** (Estimator-based Quantum Neural Network) — same circuit as VQC
  but trained against a Z-expectation observable on the last qubit and
  wrapped in `NeuralNetworkClassifier`.

Same N=200, K=6 setup as QSVM for direct comparability.

```python
%matplotlib inline
import warnings

import pandas as pd

from qml_healthcare.pipeline import run_bonus

warnings.filterwarnings("ignore")
```

```python
metrics = run_bonus()
df = pd.DataFrame(metrics).T.sort_values("roc_auc", ascending=False)
df.style.format("{:.3f}")
```
